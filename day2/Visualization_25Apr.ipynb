```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.array([12,23,34,42,51,57,63,68,74,79])
b = np.array([3,15,21,38,43,46,54,59,65,73])
c = np.array([87,79,74,65,58,51,45,36,32,27])
```

#### Line Plot

```python
plt.plot(a, b);
```

```python
plt.plot(a, c);
```

```python
plt.title('Line Plot')
plt.xlabel('first value')
plt.ylabel('Second value')
plt.xticks(range(0,90,5))
plt.yticks(range(0,90,5))
plt.xlim(0,90)
plt.ylim(0,90)
plt.plot(a, b, marker = 'o', mfc = 'r', 
         ls = '--');
plt.grid()
# marker => <, >, ^, o, H, *
```

```python
plt.title('Line Plot')
plt.xlabel('first value')
plt.ylabel('Second value')
plt.xticks(range(0,90,5))
plt.yticks(range(0,90,5))
plt.xlim(0,90)
plt.ylim(0,90)
plt.plot(a, c, marker = 'o', mfc = 'k', 
         ls = '--', color = 'r');
plt.grid()
```

```python
plt.title('Line Plot')
plt.xlabel('first value')
plt.ylabel('Second value')
plt.xticks(range(0,90,5))
plt.yticks(range(0,90,5))
plt.xlim(0,90)
plt.ylim(0,90)
plt.plot(a, b, marker = 'o', mfc = 'r', 
         ls = '--', label = 2023);
plt.plot(a, c, marker = 'o', mfc = 'k', 
         ls = '--', color = 'cyan', label = 2024);

plt.grid()
plt.legend();
```

#### subplot

```python
plt.figure(figsize=(16,10))

plt.subplot(2,2,1)
plt.title('Line Plot')
plt.xlabel('first value')
plt.ylabel('Second value')
plt.xticks(range(0,90,5))
plt.yticks(range(0,90,5))
plt.xlim(0,90)
plt.ylim(0,90)
plt.plot(a, b, marker = 'o', mfc = 'r', 
         ls = '--');
plt.grid()

plt.subplot(2,2,2)
plt.title('Line Plot')
plt.xlabel('first value')
plt.ylabel('Second value')
plt.xticks(range(0,90,5))
plt.yticks(range(0,90,5))
plt.xlim(0,90)
plt.ylim(0,90)
plt.plot(a, c, marker = 'o', mfc = 'k', 
         ls = '--', color = 'r');
plt.grid()
```

### Bar plot

```python
df1 = pd.read_csv('student.csv')
df1
```

```python
plt.title('Bar plot')
plt.xlabel('Student Name')
plt.ylabel('Marks')
plt.bar(df1['name'], df1['marks'], color = 'black',
       edgecolor = 'red', linewidth = 3);
plt.grid()
```

```python
df1['age'].value_counts()
```

```python
df1['age'].value_counts().index
```

```python
plt.xlabel('Age')
plt.ylabel('count')
plt.title('Age vs Count')
plt.bar(df1['age'].value_counts().index, 
        df1['age'].value_counts(),
       color = 'orange');
```

```python
plt.xlabel('Age')
plt.ylabel('count')
plt.title('Age vs Count')
plt.barh(df1['age'].value_counts().index, 
        df1['age'].value_counts(),
       color = 'orange');
```
